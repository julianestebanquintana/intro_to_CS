```python
import random

def dado():
  """Retorna un entero entre 1 y 6"""
  return random.choice([1,2,3,4,5,6])

def juegaDado(n):
  resultado = ''
  for i in range(n):
    resultado = resultado + str(dado())
  print(resultado)
```

```python
juegaDado(10)
```

```python
def tirarMoneda(veces):
    caras = 0.0
    for vez in range(veces):
        if random.random() < 0.5:
            caras += 1
    return caras/veces

def simulaMoneda(monedasPorTurno, Turnos):
    proporcionCaras = []
    for Turno in range(Turnos):
        proporcionCaras.append(tirarMoneda(monedasPorTurno))
    promedio = sum(proporcionCaras)/len(proporcionCaras)
    return promedio
```

```python
simulaMoneda(100,1)
```

```python
simulaMoneda(100,1)
```

```python
simulaMoneda(100,1)
```

```python
simulaMoneda(100,100)
```

```python
simulaMoneda(100,100)
```

```python
simulaMoneda(100,100000)
```

```python
vals = [1,200]
for i in range(1000):
    num1 = random.choice(range(1,100))
    num2 = random.choice(range(1,100))
    vals.append(num1+num2)
```

```python
import pylab
pylab.hist(vals, bins = 10)
```

```python
def clear(n, p, steps):
    """Asume que n y que steps, son enteros positivos; y que p es un float
        n: número inicial de moléculas
        p: probabilidad de la molécula de ser eliminada - cleared.
        steps: duración de la simulación"""
    quedan = [n]
    for t in range(steps):
        quedan.append(n*((1-p)**t))
    pylab.plot(quedan)
    pylab.xlabel('Tiempo')
    pylab.ylabel('Moléculas restantes')
    pylab.title('Eliminación del medicamento')
```

```python
clear(1000,0.01,1000)
```
